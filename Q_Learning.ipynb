```python
import numpy as np
import gym
import random
```

```python
env = gym.make("FrozenLake-v1")
```

```python
action_size = env.action_space.n
state_size = env.observation_space.n
```

```python
qtable = np.zeros((state_size, action_size))
print(qtable)
```

```python
total_episodes = 15000        
learning_rate = 0.8           
max_steps = 99                
gamma = 0.95                  


epsilon = 1.0                 
max_epsilon = 1.0             
min_epsilon = 0.01            
decay_rate = 0.005             
```

```python

import numpy as np
import random



rewards = []


for episode in range(total_episodes):
    
    state_tuple = env.reset()
    state = state_tuple[0]  
    done = False
    total_rewards = 0

    for step in range(max_steps):
        
        exp_exp_tradeoff = random.uniform(0, 1)

        
        if exp_exp_tradeoff > epsilon:
            action = np.argmax(qtable[int(state), :])  

        
        else:
            action = env.action_space.sample()

        
        new_state, reward, done, info = env.step(action)[:4]
        qtable[int(state), action] = qtable[int(state), action] + learning_rate * (
                    reward + gamma * np.max(qtable[int(new_state), :]) - qtable[int(state), action])  

        total_rewards += reward

        
        state = new_state

        
        if done:
            break

    
    epsilon = min_epsilon + (max_epsilon - min_epsilon) * np.exp(-decay_rate * episode)
    rewards.append(total_rewards)

print("Score over time: " + str(sum(rewards) / total_episodes))
print(qtable)
```
